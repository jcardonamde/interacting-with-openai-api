## Carga de modelo con fine-tuning

````python
from openai import OpenAI
client = OpenAI(api_key='INSERTA TU API KEY')

client.fine_tuning.jobs.retrieve("ftjob-iQn5nOBsOQVznpZwNXpd3ASE")
````

## Obtener archivo de resultados de fine-tuning de modelo

````python
content = client.files.content('file-3jhAYjyekcc39vvcYgPvob')

content.text
````

````python
import base64

# Asumiendo que 'content.text' contiene el texto codificado en Base64
encoded_content = content.text

# Decodificar el contenido de Base64
decoded_content = base64.b64decode(encoded_content).decode('utf-8')

print(decoded_content)
````

## Interpretación de resultados

````python
import pandas as pd
````

````python
metrics_str = decoded_content
````

````python
metrics_list = [line.split(',') for line in metrics_str.split('\n')]
````

````python
df = pd.DataFrame(metrics_list[1:], columns=metrics_list[0])
df = df.apply(pd.to_numeric, errors='coerce')
df.tail()
````

````python
#conversión a numéricos para graficación
df = df.apply(pd.to_numeric, errors='coerce')
df.tail()
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))
plt.plot(df['step'], df['train_accuracy'])
plt.title('Training Accuracy over Steps')
plt.xlabel('Step')
plt.ylabel('Training Accuracy')
plt.show()
````

````python
plt.figure(figsize=(7,4))
plt.plot(df['step'], df['train_loss'])
plt.title('Training Loss over Steps')
plt.xlabel('Step')
plt.ylabel('Training Accuracy')
plt.show()
````
